```python
import shapely
from shapely.geometry import Point
from geopandas import GeoDataFrame
```

```python
# Import packages for reprojection 
from pyproj import Proj, transform
# Set input (original) coordinate reference system (EPSG:4326 refers to Longitude and Latitude)
inProj = Proj(init='epsg:4326')
# Set output coordinate reference system (EPSG:27700 refers to British National Grid)
outProj = Proj(init='epsg:27700')
```

```python
# Create a function for conversion 
def epsg27700(row):
    return pd.Series(transform(inProj, outProj, row["lon"], row["lat"]))
```

```python
# Reproject from 'epsg:4326' to 'epsg:27700'
convert = bike_hire.apply(epsg27700,axis=1)
```

```python
convert.head()
```

```python
# Rename the column
convert = convert.rename(columns={0: 'X', 1: 'Y'})
convert.head()
```

```python
# Append the new coordinates to the crime file
bike_hire['X']=convert['X']
bike_hire['Y']=convert['Y']
```

```python
geometry = [Point(xy) for xy in zip(bike_hire.X, bike_hire.Y)]
crs = {'init': 'epsg:27700'}
gdf = GeoDataFrame(bike_hire, crs=crs, geometry=geometry)
```
